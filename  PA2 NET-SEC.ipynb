```python
def dec_bin(k): # To convert Binary string to Decimal String
    i=0
    b=0
    while True:
        b+=(k%2)*(10**i)
        k=int(k/2)
        i+=1
        if(k==0):
            break
    return b      

def xor(a,b):   # XOR of two Binary Strings
    k=[]
    for j in range(len(a)):
        if(a[j]==b[j]):
            k.append('0')
        else:
            k.append('1')
    return k        

def sub_keys(l_key,r_key):
    key0=list(l_key)+list(r_key)
    key=[key0[ii-1] for ii in key_per2]
    return key
```

```python
def hamming_distance(s1,s2):  #To Calculated Hamming Distance between two binary strings
    cnt=0
    for i in range(len(s1)):
        if(s1[i]!=s2[i]):
            cnt+=1
    return cnt        
```

```python
def round(l_plain1,r_plain1,key):
    r_exp=[r_plain1[ii-1] for ii in expan] # R Expansion Permution 32->48
    r=xor(r_exp,key) # R xor with SubKey
    r_s=[]
    for ii in range(8): # S-boxes 48->32
        k=r[(ii*6):(ii+1)*6]
        row=int(k[0])*2+int(k[5])
        col=int(k[1])*8+int(k[2])*4+int(k[3])*2+int(k[4])
        b=dec_bin(s_box[ii][row][col])
        b=str(b)
        while True:
            if(len(b)==4):
                break
            else:
                b='0'+b
        for j in b:
            r_s.append(j)  
            
    r_sp=[r_s[ii-1] for ii in per] # R Permutation
    r_f=xor(l_plain1,r_sp) # Ri = L(i-1)^F(R(i-1),Ki) 
    l_f=r_plain1 # Li = R(i-1)
    return l_f,r_f
```

```python
def key_s(l_key,r_key,n): 
    if(key_shift[n]==1):
        l_key_shif=[ii for ii in l_key[1:]]
        l_key_shif.append(l_key[0])
        r_key_shif=[ii for ii in r_key[1:]]
        r_key_shif.append(r_key[0])
    else:
        l_key_shif=[ii for ii in l_key[2:]]
        l_key_shif.append(l_key[0])
        l_key_shif.append(l_key[1])
        r_key_shif=[ii for ii in r_key[2:]]
        r_key_shif.append(r_key[0])
        r_key_shif.append(r_key[1])
    return l_key_shif,r_key_shif    
```

```python
def encrypt_rounds(plain,key):
    plain=plain.upper()
    plainbit= str(bin(int(plain, 16)))[2:].zfill(64)
    plainbit= [i for i in plainbit]
    plainbit=np.array(plainbit)
    key=key.upper()
    keybit= str(bin(int(key, 16)))[2:].zfill(64)
    keybit= [i for i in keybit]
    keybit=np.array(keybit)
    plain_per=[plainbit[i-1] for i in ini_per] # Initial Permutation in plaintext
    key_per=[keybit[i-1] for i in key_per1]
    l_plain=plain_per[:32]
    r_plain=plain_per[32:]
    l_k=key_per[:28]
    r_k=key_per[28:]
    round_data=[]
    round_data.append("".join(l_plain) + "".join(r_plain))
    for i in range(16):
        l_k,r_k=key_s(l_k,r_k,i)
        sub_key=sub_keys(l_k,r_k)
        l_plain,r_plain=round(l_plain,r_plain,sub_key)
        round_data.append("".join(l_plain) + "".join(r_plain))
    
    cy_swap=list(r_plain)+list(l_plain)
    cipher=[cy_swap[i-1] for i in final_per]
    
    return "".join(cipher) , round_data
```

## Part-1:- 5 Different Pairs of PlainText

Hamming Distace for Each Pair of PlainText = 1

i) Plaintexts:- AB3456FA7BF07D12 , AB3456FA79F07D12  
   Key:- 2247BA5687FE09CE
   
ii) Plaintexts:-13BEF5C5C6EFF098 , 13BEF5CDC6EFF098

iii) Plaintexts:- BD657AC12CFE89AC , BD657AC12CFE8DAC  
   Key:- 2247BA5687FE09CE

```python
k1='2247BA5687FE09CE'
p1_list= [['AB3456FA7BF07D12','AB3456FA79F07D12'],['13BEF5C5C6EFF098','13BEF5CDC6EFF098'],
          ['BD657AC12CFE89AC','BD657AC12CFE8DAC'],['4092BFE671498ACD','4892BFE671498ACD'],
          ['BBD456CAFE90FE5C','BBD456DAFE90FE5C']]
```

```python
perround_hamdist1=[]
for i in range(5):
    _,per_rounddata1= encrypt_rounds(p1_list[i][0],k1)
    _,per_rounddata2= encrypt_rounds(p1_list[i][1],k1)
    hamdist=[]
    for j in range(17):
        hamdist.append(hamming_distance(per_rounddata1[j],per_rounddata2[j]))
    perround_hamdist1.append(hamdist)    
```

```python
part1= pd.DataFrame({'P1':np.array(perround_hamdist1[0]),'P2':np.array(perround_hamdist1[1]),
                     'P3':np.array(perround_hamdist1[2]),'P4':np.array(perround_hamdist1[3]),
                     'P5':np.array(perround_hamdist1[4])},index=np.linspace(0,16,num=17))
```

```python
plt.figure(figsize=(14,7))
sns.set_style("whitegrid")
p=sns.lineplot(data=part1) 
p.set_xticks(range(17))
p.set_xlabel("Round number", fontsize = 20)
p.set_ylabel("Hamming Distance", fontsize = 20)
p.set_title('5 Different Pairs of Plaintext with same Key:- Hamming Distance v/s Round No', fontsize = 16)
```
